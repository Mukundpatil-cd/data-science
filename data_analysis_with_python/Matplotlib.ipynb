# Matplotlib 
Matplotlib is a Python library for data visualization that allows you to create static, animated, and interactive plots like line graphs, bar charts, scatter plots, and more.


```python
## importing matplotlib 
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

### creating a line plot

plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("---line plot---")
plt.show()
```

```python
plt.plot(x,y,color="red",linestyle="--")
```
