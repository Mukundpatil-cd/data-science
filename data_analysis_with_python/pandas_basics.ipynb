# Handling Missing Values

Missing values are common in real-world datasets. Pandas provides several methods to detect and handle them.

````python
### Check which columns have missing values
df.isnull().any()
````

````python
### Count missing values in each column
df.isnull().sum()
````

````python
### Fill missing values with a constant
df_filled = df.fillna(0)
df_filled.head()
````

````python
### Fill missing values with column mean
# Choose a numeric column that might have missing values
# Replace "POSTALCODE" with a column name from your data
if 'POSTALCODE' in df.columns:
    df['POSTALCODE_filled'] = df['POSTALCODE'].fillna(df['POSTALCODE'].mean())
    print("Column filled with mean value")
else:
    print("Column not found, replace POSTALCODE with a column that exists in your data")
    # Alternative example with any numeric column that exists
    numeric_cols = df.select_dtypes(include=['number']).columns
    if len(numeric_cols) > 0:
        sample_col = numeric_cols[0]
        df[f'{sample_col}_filled'] = df[sample_col].fillna(df[sample_col].mean())
        print(f"Example using {sample_col} instead")
````

````python
### Filling values in-place (modifying the original DataFrame)
# Choose a column that might have missing values
# This is the correct way to use inplace=True (don't try to assign it to another variable)
numeric_cols = df.select_dtypes(include=['number']).columns
if len(numeric_cols) > 0:
    sample_col = numeric_cols[0]
    # Create a copy to avoid modifying the original
    df_copy = df.copy()
    df_copy[sample_col].fillna(df_copy[sample_col].mean(), inplace=True)
    print(f"Missing values in {sample_col} filled in-place")
    print(f"NaN count before: {df[sample_col].isna().sum()}")
    print(f"NaN count after: {df_copy[sample_col].isna().sum()}")
````

## Note about inplace=True

When using `inplace=True`, the method modifies the DataFrame directly and returns `None`. You should never try to assign the result of an operation with `inplace=True` to a variable, as it will always be `None`.

**Incorrect:**
```python
df['column_name'] = df['column_name'].fillna(value, inplace=True)  # This assigns None!
```

**Correct:**
```python
df['column_name'].fillna(value, inplace=True)  # This modifies df in-place
```

Or, create a new column without using inplace:
```python
df['new_column_name'] = df['column_name'].fillna(value)  # This creates a new column
```
