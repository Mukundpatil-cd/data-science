# Pandas in python
Pandas is a Python library used for data analysis and data manipulation.

It provides fast, flexible, and expressive data structures — mainly Series (1D) and DataFrame (2D) — to work with structured data such as tables, spreadsheets, or databases.

It allows you to load, clean, transform, analyze, and visualize data easily from multiple formats like CSV, Excel, JSON, and SQL.

```python
### Merging the dataframe and joining
#1 merging the dataframe by key column
df1=pd.DataFrame({'key' : ['a','b'] , 'value1' : [1,2] })
df2=pd.DataFrame({'key' : ['a','b','c'] , 'value2' : [3,4,5]})
pd.merge(df1,df2 ,on='key',how='inner')
```
