```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,5))
for i, col in enumerate(['math score','reading score','writing score']):
    plt.subplot(1,3,i+1)
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f"Distribution of {col}")
plt.show()
```

```python

import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,3, figsize=(18,5))

axes[0].hist(df['math score'], bins=15, color="skyblue", edgecolor="black")
axes[0].set_title("Math Score Distribution")

axes[1].hist(df['reading score'], bins=15, color="lightgreen", edgecolor="black")
axes[1].set_title("Reading Score Distribution")

axes[2].hist(df['writing score'], bins=15, color="salmon", edgecolor="black")
axes[2].set_title("Writing Score Distribution")

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.barplot(x="gender", y="math score", data=df, ci=None)
plt.title("Average Math Score by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="lunch", y="math score", data=df, ci=None)
plt.title("Math Score by Lunch Type")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="parental level of education", y="reading score", data=df, ci=None)
plt.xticks(rotation=45)
plt.title("Reading Score by Parental Education")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation between Scores")
plt.show()
```
